```python
# Problem 1
# Initialize variables
num = 150
count = 0
sum_odd = 0
product = 1
sum_square = 0

while num <= 1000:
    if num % 2 != 0:   # Check if number is odd
        count += 1
        sum_odd += num
        product *= num
        sum_square += num ** 2
    num += 1

# Display results
print("Count of odd numbers:", count)
print("Sum of odd numbers:", sum_odd)
print("Product of odd numbers:", product)
print("Sum of squares of odd numbers:", sum_square)
```

•	Start the number at 150 and run a while loop until 1000 (inclusive).

•	Create variables before the loop to store:

*   count → number of odd numbers
*   sum_odd → sum of odd numbers
*   product → multiplication of odd numbers
*   sum_square → sum of squares of odd numbers

•	Inside the loop, check if a number is odd using num % 2 != 0.

•	If the number is odd:
*   Increase count by 1
*   Add the number to sum_odd
*   Multiply it to product
*   Add its square (num ** 2) to sum_square

•	Increase the number by 1 each time (num += 1) to avoid an infinite loop.

•	After the loop finishes, print all the calculated results.


```python
# Problem 2
# Initialize variables
count = 0
sum_odd = 0
product = 1
sum_square = 0

# Loop from 150 to 1000 (inclusive)
for num in range(150, 1001):
    if num % 2 != 0:   # Check if number is odd
        count += 1
        sum_odd += num
        product *= num
        sum_square += num ** 2

# Display results
print("Count of odd numbers:", count)
print("Sum of odd numbers:", sum_odd)
print("Product of odd numbers:", product)
print("Sum of squares of odd numbers:", sum_square)
```

I used a for loop because the starting value (150) and ending value (1000) are known.

I used range(150, 1001) to include 1000, since the range function does not include the last number.

Before the loop, I initialized four variables:

*   count = 0 to count odd numbers
*   sum_odd = 0 to store the sum
*   product = 1 to store the multiplication result
*   sum_square = 0 to store the sum of squares

Inside the loop, I checked if a number is odd using num % 2 != 0.

If the number is odd:

*   Increase count by 1
*   Add the number to sum_odd
*   Multiply it to product
*   Add its square (num ** 2) to sum_square

After the loop finishes, the final results are printed.

```python
# Problem 3
# Input value of M
M = int(input("Enter value of M: "))

# Initialize variables
k = 1
s = 0

# While loop
while k <= M:
    s = s + (1 / k)
    k = k + 1

# Output result
print("The sum is:", s)
```

The program calculates the sum s = 1 + 1/2 + 1/3 + … + 1/M (harmonic sum).

I start by setting k = 1 because the summation begins at 1.

I initialize s = 0 to store the total sum.

I use a while loop that runs as long as k <= M.

Inside the loop:

*   Add 1/k to s.
*   Increase k by 1 (k = k + 1).

The loop stops when k becomes greater than M.

Finally, the program prints the computed sum.

```python
# Problem 4
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break

    try:
        number = float(user_input)
        total += number
        count += 1
    except:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
else:
    average = 0

print(total, count, average)
```

total stores the sum of valid numbers.

count keeps track of how many valid numbers were entered.

The while True loop keeps running until the program encounters the break statement when the user enters "done".

Inside try, the program attempts to convert input into a number using float().

If conversion fails, the except block prints "Invalid input" and skips that entry.

After the loop ends, the average is calculated as total / count.

Finally, the program prints the total, count, and average.

```python
# Problem 5
max_num = None
min_num = None

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break

    try:
        number = float(user_input)

        if max_num is None or number > max_num:
            max_num = number

        if min_num is None or number < min_num:
            min_num = number

    except:
        print("Invalid input")
        continue

if max_num is not None and min_num is not None:
    print("Maximum:", max_num)
    print("Minimum:", min_num)
else:
    print("No valid numbers were entered.")
```

Initialize max_num and min_num to None, then use a while True loop to repeatedly ask the user for input.
If the user enters "done", exit the loop using break.
Use a try/except block to attempt to convert the input to a number using float(). If the conversion fails, the except block handles the error.

If valid:

*   Update max_num if the number is greater than the current maximum.
*   Update min_num if the number is smaller than the current minimum.

If the input is invalid, print an error message and continue to the next iteration.

After the loop, print the maximum and minimum values.

```python
# Problem 7
n = 3
for x in [2, 5, 8]:
    n = n + x
    print(n)
```

The program starts by setting n = 3.

The for loop goes through each number in the list [2, 5, 8].

In each iteration, the program adds the current value of x to n using n = n + x.

Step-by-step updates:

First: 3 + 2 = 5 → prints 5

Second: 5 + 5 = 10 → prints 10

Third: 10 + 8 = 18 → prints 18

Since print(n) is inside the loop, each updated value is printed during the iteration.
